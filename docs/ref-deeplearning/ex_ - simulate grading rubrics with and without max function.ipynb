```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Update number of students
n_students = 1000

# Simulate grades for a larger group of students
homework_grades = np.random.uniform(0, 14, n_students)  # Homework scores out of 14
midterm_grades = np.random.uniform(0, 100, n_students)  # Midterm scores out of 100
final_grades = np.random.uniform(0, 100, n_students)  # Final scores out of 100

# Recalculate exam and total scores for the new student count
exam_scores = np.maximum(final_grades, 0.55 * final_grades + 0.45 * midterm_grades)
total_grades_max = np.maximum(0.86 * exam_scores + 0.14 * homework_grades, exam_scores)
total_grades_no_max = 0.43 * midterm_grades + 0.43 * final_grades + 0.14 * homework_grades
```

```python
# Plot distribution for 1000 students
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(total_grades_max, bins=20, alpha=0.7, color='blue')
plt.title('Grade Distribution with Max Formula (1000 Students)')
plt.xlabel('Grade')
plt.ylabel('Number of Students')

plt.subplot(1, 2, 2)
plt.hist(total_grades_no_max, bins=20, alpha=0.7, color='green')
plt.title('Grade Distribution without Max Formula (1000 Students)')
plt.xlabel('Grade')
plt.ylabel('Number of Students')

plt.tight_layout()
plt.show()
```
